# **Now We Will Do Text Preprocessing**<br>
and follow will these essential steps.
- Convert the all words lower case.
- Tokenizaton
- Remove the special characters.
- Remove punctuation
- Stemming

```python
import string
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# We will Create a function that will perform the above task

def text_transform(text):
  text = text.lower()
  text = nltk.word_tokenize(text)

  y = []

  for i in text:
    if i.isalnum():
      y.append(i)

  text = y[:]
  y.clear()

  for i in text:
    if i not in stopwords.words('english') and i not in string.punctuation:
      y.append(i)

  text = y[:]
  y.clear()

  for i in text:
    y.append(ps.stem(i))

  return " ".join(y)
```

# Check the function result

```python
text_transform('Hi How Are %% / Pakistan is a beautiful ciuntry? ')
```

```python
Email_data['text_Transformed'] = Email_data['messages'].apply(text_transform)
Email_data.head()
```

#  <font color='bluewight' ><strong>Training with different Model</font>

```python
Gnb = GaussianNB()
Mnb = MultinomialNB()
Rfc = RandomForestClassifier()
KnbC = KNeighborsClassifier()
DtC = DecisionTreeClassifier()
Xgb = XGBClassifier()
```

```python
from sklearn.metrics import accuracy_score,f1_score, confusion_matrix, precision_score
```

```python
#  Training with GaussianNB Model
gnb_X_train = X_train
gnb_X_test =  X_test
gnb_y_train = y_train
gnb_y_test = y_test

Gnb.fit(gnb_X_train, gnb_y_train)

Gnb_pred = Gnb.predict(gnb_X_test)

#  Training with MultinomialNB Model
Mnb_X_train = X_train
Mnb_X_test =  X_test
Mnb_y_train = y_train
Mnb_y_test = y_test

Mnb.fit(Mnb_X_train, Mnb_y_train)

Mnb_pred = Mnb.predict(Mnb_X_test)

#  Training with  RandomForest Model
Rfc_X_train = X_train
Rfc_X_test =  X_test
Rfc_y_train = y_train
Rfc_y_test = y_test

Rfc.fit(Rfc_X_train, Rfc_y_train)

Rfc_pred = Rfc.predict(Rfc_X_test)

#  Training with  KNeighbors Model
Kn_X_train = X_train
Kn_X_test =  X_test
Kn_y_train = y_train
Kn_y_test = y_test

KnbC.fit(Kn_X_train, Kn_y_train)

Kn_pred = KnbC.predict(Kn_X_test)

#  Training with  DecisionTree Model
Dt_X_train = X_train
Dt_X_test =  X_test
Dt_y_train = y_train
Dt_y_test = y_test

DtC.fit(Dt_X_train, Dt_y_train)

Dt_pred = DtC.predict(Dt_X_test)

#  Training with  XGBoost Model
Xgb_X_train = X_train
Xgb_X_test =  X_test
Xgb_y_train = y_train
Xgb_y_test = y_test

Xgb.fit(Xgb_X_train, Xgb_y_train)

Xgb_pred = Xgb.predict(Xgb_X_test)
```

```python
ps_gnb = precision_score(gnb_y_test,Gnb_pred)
as_gnb = accuracy_score(gnb_y_test,Gnb_pred)
fs_gnb = f1_score(gnb_y_test,Gnb_pred)

ps_mnb = precision_score(Mnb_y_test,Mnb_pred)
as_mnb = accuracy_score(Mnb_y_test,Mnb_pred)
fs_mnb = f1_score(Mnb_y_test,Mnb_pred)

ps_rf = precision_score(Rfc_y_test,Rfc_pred)
as_rf = accuracy_score(Rfc_y_test,Rfc_pred)
fs_rf = f1_score(Rfc_y_test,Rfc_pred)

ps_kn = precision_score(Kn_y_test,Kn_pred)
as_kn = accuracy_score(Kn_y_test,Kn_pred)
fs_kn = f1_score(Kn_y_test,Kn_pred)

ps_dt = precision_score(Dt_y_test,Dt_pred)
as_dt = accuracy_score(Dt_y_test,Dt_pred)
fs_dt = f1_score(Dt_y_test,Dt_pred)

ps_xg = precision_score(Xgb_y_test,Xgb_pred)
as_xg = accuracy_score(Xgb_y_test,Xgb_pred)
fs_xg = f1_score(Xgb_y_test,Xgb_pred)
```

```python
E_data = {
    'Model': ['GNB', 'MNB', 'RF', 'KN', 'DT','XGBoost'],
    'PrecisionScore': [ps_gnb, ps_mnb, ps_rf, ps_kn, ps_dt, ps_xg],
    'AccScore': [as_gnb, as_mnb, as_rf, as_kn, as_dt, as_xg],
    'f1_Score': [fs_gnb,fs_mnb, fs_rf, fs_kn, fs_dt, fs_xg]
    }

e_df = pd.DataFrame(E_data)
print(e_df)
```
